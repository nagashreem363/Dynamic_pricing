```python
import matplotlib.pyplot as plt
import seaborn as sns

season_sales = data1.groupby('Season')['PRICE_CURRENT'].sum().reset_index()

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Season', y='PRICE_CURRENT', data=season_sales, palette='viridis') # Change palette for different colors
plt.title('Total Sales by Season')
plt.xlabel('Season')
plt.ylabel('Total Sales')

# Add numbers above the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f') # Format the numbers

plt.show()

highest_sales_season = season_sales.loc[season_sales['PRICE_CURRENT'].idxmax()]
print(f"The season with the highest sales is: {highest_sales_season['Season']}")
print(f"Total sales in this season: {highest_sales_season['PRICE_CURRENT']:.2f}")
```

```python
fall_data = data1[data1['Season'] == 'Fall']

# Group by Category and sum the sales
category_sales_fall = fall_data.groupby('CATEGORY')['PRICE_CURRENT'].sum().reset_index()

# Find the category with the highest sales in Fall
highest_category_fall = category_sales_fall.loc[category_sales_fall['PRICE_CURRENT'].idxmax()]
print(f"In the Fall season, the category with the highest sales is: {highest_category_fall['CATEGORY']}")
print(f"Total sales in this category: {highest_category_fall['PRICE_CURRENT']:.2f}")

# Group by PRODUCT_NAME within the Fall season and sum the sales
product_sales_fall = fall_data.groupby('PRODUCT_NAME')['PRICE_CURRENT'].sum().reset_index()

# Find the product with the highest sales in Fall
highest_product_fall = product_sales_fall.loc[product_sales_fall['PRICE_CURRENT'].idxmax()]
print(f"In the Fall season, the product with the highest sales is: {highest_product_fall['PRODUCT_NAME']}")
print(f"Total sales for this product: {highest_product_fall['PRICE_CURRENT']:.2f}")


# Plot the sales for the top N categories in Fall
top_n_categories = 10 # You can adjust this number
top_categories_fall = category_sales_fall.nlargest(top_n_categories, 'PRICE_CURRENT')

plt.figure(figsize=(12, 8))
ax = sns.barplot(x='CATEGORY', y='PRICE_CURRENT', data=top_categories_fall, palette='viridis')
plt.title(f'Top {top_n_categories} Categories with Highest Sales in Fall')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout() # Adjust layout to prevent labels overlapping

# Add numbers above the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f') # Format the numbers

plt.show()


# Plot the sales for the top N products in Fall
top_n_products = 10 # You can adjust this number
top_products_fall = product_sales_fall.nlargest(top_n_products, 'PRICE_CURRENT')

plt.figure(figsize=(12, 8))
ax = sns.barplot(x='PRODUCT_NAME', y='PRICE_CURRENT', data=top_products_fall, palette='viridis')
plt.title(f'Top {top_n_products} Products with Highest Sales in Fall')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout() # Adjust layout to prevent labels overlapping

# Add numbers above the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f') # Format the numbers


plt.show()
```

```python


# Group by 'CATEGORY' and sum 'PRICE_CURRENT' to get total sales per category
category_sales = data1.groupby('CATEGORY')['PRICE_CURRENT'].sum().reset_index()

# Sort in descending order and get the top category
highest_sales_category = category_sales.sort_values(by='PRICE_CURRENT', ascending=False).iloc[0]

print(f"The category with the highest sales is: {highest_sales_category['CATEGORY']}")
print(f"Total sales in this category: {highest_sales_category['PRICE_CURRENT']:.2f}")


top_n = 10
top_categories = category_sales.sort_values(by='PRICE_CURRENT', ascending=False).head(top_n)

plt.figure(figsize=(12, 7))
sns.barplot(x='PRICE_CURRENT', y='CATEGORY', data=top_categories, palette='viridis')
plt.title(f'Top {top_n} Categories by Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Category')
plt.show()
```

```python
# prompt: shipping locationa and their total sales graph

# Group by SHIPPING_LOCATION and sum the sales
location_sales = data1.groupby('SHIPPING_LOCATION')['PRICE_CURRENT'].sum().reset_index()

plt.figure(figsize=(12, 7))
ax = sns.barplot(x='SHIPPING_LOCATION', y='PRICE_CURRENT', data=location_sales, palette='viridis')
plt.title('Total Sales by Shipping Location')
plt.xlabel('Shipping Location')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout() # Adjust layout to prevent labels overlapping

# Add numbers above the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f') # Format the numbers

plt.show()
```

```python
# prompt: more products saled in 10001 location graph

# Filter data for SHIPPING_LOCATION 10001
location_10001_data = data1[data1['SHIPPING_LOCATION'] == 10001]

# Group by PRODUCT_NAME and sum the sales for location 10001
product_sales_10001 = location_10001_data.groupby('PRODUCT_NAME')['PRICE_CURRENT'].sum().reset_index()

# Sort by sales in descending order and get the top products
top_n_products_10001 = 10 # Adjust as needed
top_products_by_sales_10001 = product_sales_10001.sort_values(by='PRICE_CURRENT', ascending=False).head(top_n_products_10001)

# Create the bar chart
plt.figure(figsize=(12, 8))
ax = sns.barplot(x='PRICE_CURRENT', y='PRODUCT_NAME', data=top_products_by_sales_10001, palette='viridis')
plt.title(f'Top {top_n_products_10001} Products Sold in Location 10001')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')

# Add numbers next to the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')

plt.show()
```
